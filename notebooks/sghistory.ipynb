```python
import requests
from bs4 import BeautifulSoup

def load_xml_from_url(url):
    response = requests.get(url)
    return response.text

def extract_urls(xml_data):
    soup = BeautifulSoup(xml_data, 'lxml')
    urls = [element.text for element in soup.find_all('loc')]
    return urls

# Example usage:
xml_data = load_xml_from_url('https://eresources.nlb.gov.sg/history/sitemap.xml')
urls = extract_urls(xml_data)
print(urls)
print(len(urls))
```
